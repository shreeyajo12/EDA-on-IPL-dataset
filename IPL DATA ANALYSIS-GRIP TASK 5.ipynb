## Merging both the datasets for insights

```python
merge = pd.merge(deliveries,data,left_on='match_id', right_on ='id')
merge.head()
```

```python
merge.describe()
```

```python
merge.head(2)
```

## Top 10 batsmen with the highest runs

```python
bats=merge.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).plot.bar(figsize=(15,8),color='orange',ylabel='Total runs')
plt.title("TOP 10 BATSMEN")
```

## SIXES HIT BY BATSMEN

```python
six=merge[merge['batsman_runs']==6]
c=six.groupby('batsman').count()['batsman_runs'].sort_values(ascending=False).head(10)
c
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=merge,x=c.index,y=c)
```

### CH GAYLE HAS SECURED THE MOST NUMBER OF SIXES

## TOP BOWLERS

```python
bow=merge.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10).plot.bar(figsize=(15,8),color='purple',ylabel='Total wickets')
```

## SL MALINGA is the top bowler in IPL as he has taken the most number of wickets than any other bowler.
